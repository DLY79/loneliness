```python
#removing duplicate info
lonely_df = lonely_df[lonely_df.GEO != "Canada (excluding territories)"]
```

```python
lonely2_df = lonely_df[lonely_df.GEO != "Atlantic Region"]
```

```python
lonely3_df = lonely2_df[lonely2_df.GEO != "Prairies Region"]
```

```python
# Removing "Total, all persons"
lonely4_df = lonely3_df[lonely3_df.Gender != "Total, all persons"]
```

```python
#Get rid of excess columns
lonely4_df=lonely4_df.drop(["TERMINATED", "STATUS", "SYMBOL", "DGUID","Statistics","VECTOR","COORDINATE", "UOM_ID", "SCALAR_FACTOR","SCALAR_ID", "DECIMALS", "REF_DATE"], axis=1)
```

```python
lonely5_df = lonely4_df[lonely4_df.UOM == 'Percent']
```
