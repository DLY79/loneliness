```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loneliness_df = pd.read_csv('loneliness.csv')
```

```python
loneliness_df.tail()
```

```python
l1_df=loneliness_df.groupby(['GEO','Gender','Indicators'])['VALUE'].mean()
```

```python
l1_df
```

```python
l1_df.to_csv('l1.csv')
```

```python
l2_df = pd.read_csv('l1.csv')
```

```python
l2_df
```

```python
names = l2_df['GEO'].unique().tolist()   #find unique values
ab = l2_df[l2_df['GEO']=='Alberta']        
bc = l2_df[l2_df['GEO']=='British Columbia']
mb = l2_df[l2_df['GEO']=='Manitoba']        
qc = l2_df[l2_df['GEO']=='Quebec']
nb = l2_df[l2_df['GEO']=='New Brunswick']        
ns = l2_df[l2_df['GEO']=='Nova Scotia']
nl = l2_df[l2_df['GEO']=='Newfoundland and Labrador']        
on = l2_df[l2_df['GEO']=='Ontario']
sk = l2_df[l2_df['GEO']=='Saskatchewan']        
pe = l2_df[l2_df['GEO']=='Prince Edward Island']
```

```python
bc
```

```python

sns.set_style('darkgrid')
sns.set_context("talk", font_scale = .5, rc={"grid.linewidth": 2})
ab_plot=sns.catplot(x = "Indicators",       # x variable name
            y = "VALUE",       # y variable name
            hue = "Gender",  # group variable name
            data = ab,     # dataframe to plot
            kind = "bar",
            order = ("Rarely or never feels lonely","Sometimes feels lonely","Always or often feels lonely"),
            
            ).set(title="Alberta")
plt.savefig('js/images/alberta.png', dpi = 300)

```
